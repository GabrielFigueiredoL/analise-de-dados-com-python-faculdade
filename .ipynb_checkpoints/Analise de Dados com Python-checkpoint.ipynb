```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import json

with open('data.json', 'r', encoding='utf-8') as file:
    data = file.read()
    
data_dict = json.loads(data)
df = pd.DataFrame(data_dict['data'])
df['selectedDate'] = pd.to_datetime(df['selectedDate'], format='%d/%m/%Y')
df['Month'] = df['selectedDate'].dt.to_period('M')
```

## Contagem de clientes por mês

```python
sales_per_month = df['Month'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
sales_per_month.plot(kind='bar', color='#89CCC5')
plt.title('Quantidade de Clientes por Mês')
plt.xlabel('Mês')
plt.ylabel('Quantidade de Clientes')
plt.xticks(range(len(sales_per_month)), [month.strftime('%m/%Y') for month in sales_per_month.index])
plt.xticks(rotation=45)
plt.show()
```

## Faturamento por mês

```python
# Gráfico de barras - Faturamento por mês
revenue_per_month = df.groupby('Month')['finalValue'].sum().sort_index()
plt.figure(figsize=(10, 5))
revenue_per_month.plot(kind='bar', color='#89CCC5')
plt.title('Faturamento por Mês')
plt.xlabel('Mês')
plt.ylabel('Faturamento (R$)')
plt.xticks(range(len(sales_per_month)), [month.strftime('%m/%Y') for month in sales_per_month.index])
plt.xticks(rotation=0)
plt.show()
```

## Quantidade total de itens vendidos por mês

```python
# Extrair os itens vendidos e a quantidade
items_data = df.explode('selectedItems')
items_data['qtd'] = items_data['selectedItems'].apply(lambda x: int(x['qtd']))
items_data['item_name'] = items_data['selectedItems'].apply(lambda x: x['item']['name'])

# Agrupar a quantidade total de itens vendidos por mês
monthly_total = items_data.groupby('Month')['qtd'].sum()

# Gráfico de barras - Quantidade total de itens vendidos por mês
plt.figure(figsize=(10, 6))
monthly_total.plot(kind='bar', color='#89CCC5')
plt.title('Quantidade Total de Itens Vendidos por Mês')
plt.xlabel('Mês')
plt.ylabel('Quantidade Vendida')
plt.xticks(range(len(monthly_total)), [month.strftime('%m/%Y') for month in monthly_total.index])
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Quantidade de cada item vendido por mês

```python
# Extrair os itens vendidos e a quantidade
items_data = df.explode('selectedItems')
items_data['qtd'] = items_data['selectedItems'].apply(lambda x: int(x['qtd']))
items_data['item_name'] = items_data['selectedItems'].apply(lambda x: x['item']['name'])

# Gráfico de barras - Quantidade de cada item vendido no mês
items_sold_per_month = items_data.groupby(['Month', 'item_name'])['qtd'].sum().unstack().fillna(0)
items_sold_per_month.plot(kind='bar', stacked=False, figsize=(12, 6))
plt.title('Quantidade de Cada Item Vendido por Mês')
plt.xlabel('Mês')
plt.ylabel('Quantidade Vendida')
plt.legend(title='Item')
plt.xticks(range(len(items_sold_per_month)), [month.strftime('%m/%Y') for month in sales_per_month.index])
plt.xticks(rotation=0)
plt.show()
```
